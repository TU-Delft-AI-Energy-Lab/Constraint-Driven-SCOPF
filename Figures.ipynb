```python
import matplotlib.pyplot as plt
import numpy as np

def calculate_num_params_MLP(input_neurons, output_neurons):
    hidden_neurons = 8
    hidden_layers = 3

    num_params_input_to_hidden = (input_neurons + 1) * hidden_neurons  # +1 for bias
    num_params_hidden_to_hidden = (hidden_neurons + 1) * hidden_neurons  # +1 for bias
    num_params_hidden_to_output = (hidden_neurons + 1) * output_neurons  # +1 for bias

    total_num_params = (num_params_input_to_hidden +
                        num_params_hidden_to_hidden * (hidden_layers - 1) +
                        num_params_hidden_to_output)

    return total_num_params


def calculate_num_params_GNN(num_features, num_of_outputs, conv_hid_features1, conv_hid_features2, K):
    num_of_outputs = 1
    hidden_params1 = (num_features + 1) * conv_hid_features1 * K  # +1 for bias
    hidden_params2 = (conv_hid_features1 + 1) * conv_hid_features2 * K  # +1 for bias
    hidden_params3 = (conv_hid_features2 + 1) * conv_hid_features2 * K  # +1 for bias
    output_params = (conv_hid_features2 + 8) * num_of_outputs  # +8 for extra_feature, +1 for bias

    total_num_params = hidden_params1 + hidden_params2 + hidden_params3 + output_params

    return 457

K = 3
num_features = 1   # Example number of input features
conv_hid_features1 = 16  # Example number of hidden neurons in the first convolutional layer
conv_hid_features2 = 16  # Example number of hidden neurons in the second convolutional layer

input_size = 16
output_sizes = list(range(0, 201, 20))
number_nodes = list(range(0, 101, 10))

num_params_MLP = [calculate_num_params_MLP(output_size, output_size) for output_size in output_sizes]
num_params_GNN = [calculate_num_params_GNN(num_features, output_size, conv_hid_features1, conv_hid_features2, K) for output_size in output_sizes]

plt.figure(figsize=(6, 3))  # Adjust the figure size as needed
plt.plot(number_nodes, num_params_MLP, marker="3", label='MLP', markersize=20, linestyle='None')
plt.plot(number_nodes, num_params_GNN, marker="4", label='GNN', markersize=20, linestyle='None')
plt.xlabel('Number of buses')
plt.ylabel('Number of parameters')

ax = plt.gca()
for spine in ['top', 'right', 'bottom', 'left']:
    ax.spines[spine].set_linewidth(0.5)
ax.yaxis.set_ticks_position('right')
ax.xaxis.set_ticks_position('top')
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
ax.tick_params(bottom=True, top=True, left=True, right=True)
ax.tick_params(labelbottom=True, labeltop=False, labelleft=True, labelright=False)

# Set custom ticks and labels
plt.xticks(number_nodes)

plt.legend()  # Add a legend to the plot
plt.tight_layout()  # Improve spacing between elements
plt.savefig('parameterization_comparison.png', dpi=300)  # Save the plot to a file
plt.show()
```
